```python
from bokeh.plotting import figure, show
from bokeh.io import output_file

x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 3]

p = figure(title="Simple Scatter Plot", x_axis_label='X-axis', y_axis_label='Y-axis')

p.circle(x, y, size=10, color="navy", alpha=0.5)

output_file("scatter_plot.html")
show(p)
```

# Q2. What are glyphs in Bokeh, and how can you add them to a Bokeh plot? Explain with an example.


# answer-2-In Bokeh, glyphs are visual markers that represent data points on a plot. Glyphs can be various shapes like circles, squares, lines, bars, patches, etc., and they are used to visually display data in different ways. Adding glyphs to a Bokeh plot allows you to customize the appearance and style of the data representation.

```python
x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 3]

p = figure(title="Glyphs Example", x_axis_label='X-axis', y_axis_label='Y-axis')

p.circle(x, y, size=10, color="navy", alpha=0.5, legend_label="Circle")
p.square(x, [i*2 for i in y], size=12, color="green", alpha=0.6, legend_label="Square")
p.line(x, y, line_width=2, color="red", alpha=0.8, legend_label="Line")
output_file("glyphs_example.html")
show(p)
```

# Q3. How can you customize the appearance of a Bokeh plot, including the axes, title, and legend?

```python
x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 3]

# Create a Bokeh figure with customizations
p = figure(title="My Custom Bokeh Plot",x_axis_label="X-axis Label",y_axis_label="Y-axis Label",
           background_fill_color="lightgrey",
           border_fill_color="white")
p.circle(x, y, size=10, color="navy", alpha=0.5, legend_label="Circle")
p.legend.title = "Legend Title"
p.legend.label_text_color = "navy"
output_file("custom_bokeh_plot.html")
show(p)
```

# Q4. What is a Bokeh server, and how can you use it to create interactive plots that can be updated in real time?

# Answer-4-Bokeh Server is a component of Bokeh, a Python library for creating interactive visualizations, that allows you to create and serve interactive plots which can be updated in real-time. The Bokeh Server enables the creation of interactive web applications where the plot updates dynamically based on user interactions or changes in the underlying data.

```python
from bokeh.plotting import figure, curdoc
from bokeh.models import ColumnDataSource
from bokeh.io import output_file, show
from random import randint
from time import sleep

x = [0]
y = [0]

p = figure(title="Real-Time Updating Plot", x_axis_label='X-axis', y_axis_label='Y-axis')
source = ColumnDataSource(data=dict(x=x, y=y))
p.line('x', 'y', source=source, line_width=2)

def update():
    x.append(x[-1] + 1)
    y.append(randint(0, 10))
    source.data = dict(x=x, y=y)

curdoc().add_periodic_callback(update, 1000)  # Update every 1000 milliseconds (1 second)
output_file("realtime_plot.html")
show(p)
# bokeh serve --show script_name.py
```

```python
from flask import Flask, render_template
from bokeh.plotting import figure
from bokeh.embed import file_html
from bokeh.resources import CDN

app = Flask(__name__)

@app.route('/')
def plot_bokeh():
    p = figure(title="Bokeh Plot", x_axis_label='X-axis', y_axis_label='Y-axis')
    p.line([1, 2, 3, 4, 5], [6, 7, 2, 4, 3], line_width=2)
    plot_html = file_html(p, CDN, "My Bokeh Plot")

    return render_template('index.html', plot_html=plot_html)

if __name__ == '__main__':
    app.run(debug=True)
```
